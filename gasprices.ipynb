```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

gas = pd.read_csv('gas_price_by_state.csv')
print(gas)

plt.figure(figsize=(20,2))

plt.bar(gas["name"], gas["gasoline"])
plt.bar(gas["name"], gas["midGrade"], bottom = gas["gasoline"])
plt.bar(gas["name"], gas["premium"], bottom = np.add(gas["gasoline"],gas["midGrade"]))

plt.xticks(rotation=90)
```
